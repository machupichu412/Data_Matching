```python
import pandas as pd
```

```python
assignment = pd.read_excel("2A PORTS ASSIGNMENT 5.28.25.xlsx")
port_map = pd.read_excel("2A Floor Port Map 5.27.25.xlsx", header=1)
```

```python
assignment.rename(columns={"PORT A ": "Port A"}, inplace=True)
assignment.shape
```

```python
# patch panel port from assignment goes to port map
port_cols = [f"Port {c}" for c in ['A', 'B', 'C', 'D', 'E', 'F']]

# replace all NaN values in port_map Patch Panel Port columns with values for assignment, matched by "Serving IDF Room" column
for idx, row in assignment.iterrows():
    # Find the matching row in port_map based on "Serving IDF Room"
    match_row = port_map[port_map["Serving IDF Room"] == row["Serving IDF Room"]]
    if not match_row.empty:
        for col in port_cols:
            new_value = assignment.at[idx, col]
            if pd.isna(new_value):
                continue
            print(f"Updating {col} for {row['Serving IDF Room']} with value {new_value}")
            port_map.loc[match_row.index, f"Patch Panel {col}"] = assignment.at[idx, col]

port_map.head()

    
```

```python
# export the updated port_map to a new Excel file
output_file = "updated_port_map_2a.xlsx"
port_map.to_excel(output_file, index=False)
print(f"Updated port map saved to {output_file}")
```
